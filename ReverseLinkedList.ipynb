# Reverse a Linked List

The algorithm to reverse a linked list looks like this:

```python
def reverse(head):
  previous = None
  current = head
  next_node = None

  while current:
    next_node = current.next
    current.next = previous
    previous = current
    current = next_node

  return previous
```

## Reverse a Sublist

A number of problems may reduce to reversing a sublist. For example: reverse k elements, reverse groups of n nodes, etc.

Say we want to reverse between p and q nodes, the algorithm looks like this:

```
1. Skip to p
2. Remember the node at position p-1
3. Reverse the nodes from p to q
4. Connect p-1 to q in the reversed list (q = previous)
5. Connect p to the next element after the reversed list (current)
```

Code:

```python
# 1 > 2 > 3 > 4 > 5, p=2, q=4
# 1 > 4 > 3 > 2 > 5
def reverse_sub_list(head, p, q):
  if p == q:
    return head

  previous = None
  current = head
  i = 0
  
  # Skip to p
  while current and i < p - 1:
    previous = current
    current = current.next
    i += 1

  # p-1 (need to connect this to first node in reversed list)
  last_node_of_first_part = previous
  # p (need to connect this node after reversed list)
  last_node_of_sub_list = current
  next = None  # will be used to temporarily store the next node

  i = 0
  # reverse nodes between 'p' and 'q'
  while current and i < q - p + 1:
    next = current.next
    current.next = previous
    previous = current
    current = next
    i += 1

  # connect with the first part (i.e. p-1)
  if last_node_of_first_part:
    # 'previous' is now the first node of the sub-list
    last_node_of_first_part.next = previous
  # this means p == 1 i.e., we are changing the first node (head) of the LinkedList
  else:
    head = previous

  # connect with the last part
  last_node_of_sub_list.next = current
  return head
```
